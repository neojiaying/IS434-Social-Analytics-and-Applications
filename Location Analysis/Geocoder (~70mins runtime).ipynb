```python
import requests, json
import pandas as pd

sgloc = pd.read_csv("sglocations2.txt")
df = pd.DataFrame(columns=["location", "latitude", "longitude"])

def geo_loc_latlong(search):
    address_list = [search + ", Singapore"]
    url = 'https://nominatim.openstreetmap.org/search'

    for queries in address_list:
        param = {'q':queries, 'format':'json', 'limit':1}
        r = requests.get(url, params=param)
        try:
            print(r.json()[0]["lat"], r.json()[0]["lon"])
        except:
            print("No lat long")
    
    try:
        return {"lat" : r.json()[0]["lat"], "lon" : r.json()[0]["lon"]}
    except:
        return {"lat": None, "lon": None}

for index, row in sgloc.iterrows():
    coordinates = (geo_loc_latlong(row.location))
    df.loc[len(df)] = [row.location, coordinates["lat"], coordinates["lon"]]
    
    
```
